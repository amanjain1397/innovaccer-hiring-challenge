```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as scipy
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```
cbd = pd.read_csv('./input/Customer Bill Detail.csv', header = 0)
coid = pd.read_csv('./input/Customer Order Item Details.csv', header = 0)
mp = pd.read_csv('./input/Market Pricing.csv', header = 0)
```

```
no_love = list(mp.loc[mp['What people love here'].isnull()].index)
love_index = list(mp.loc[mp['What people love here'].notnull()].index)
love = list(mp['What people love here'].str.split(', ', expand=True).stack().unique())
```

```
matrix = pd.DataFrame(np.zeros((len(mp), len(love))), columns = love)
for i in love_index:
  for word in mp.loc[i, 'What people love here'].split(', '):
    matrix.iloc[i, love.index(word)] = 1  
```

```
temp = []
for restaur in mp['Restaurant'].value_counts().keys().tolist():
  temp.append(list(mp.loc[mp['Restaurant'] == restaur].index)[0])
temp = np.sort(temp)
matrix = matrix.loc[temp]
```

```
print(np.array(matrix.sum().index))
print(matrix.sum().values)
```

```
matrix_df = pd.DataFrame(data = [np.array(matrix.sum().index), matrix.sum().values], index = ['Dish', 'Count']).T
matrix_df.sort_values(by = 'Count', ascending = False, inplace = True)
```

```
matrix_df['Qty Bins'] = pd.cut(matrix_df['Count'], bins = 3).astype(str)
```

```
matrix_df
```

```
matrix_df['Qty Bins'] = matrix_df['Qty Bins'].map({'(7.667, 11.0]' : 'High', '(4.333, 7.667]' : 'Medium', '(0.99, 4.333]' : 'Low'})
```

```
goli_top = coid['Dish Name'].value_counts().to_frame().reset_index()
goli_top.columns = ['Dish', 'Count']
goli_todrop = [1, 4, 8, 10, 11, 14, 16, 22, 24] #Index of StopDishes
goli_top.drop(labels = goli_todrop, axis =0, inplace = True)
goli_top['Qty Bins'] = pd.cut(goli_top['Count'], bins = 3).astype(str)
```

```
goli_top
goli_top['Qty Bins'] = goli_top['Qty Bins'].map({'(886.333, 1329.0]' : 'High', '(443.667, 886.333]' : 'Medium', '(-0.328, 443.667]' : 'Low'})
```

```
to_promote_medium = []
for dish in list(goli_top.loc[goli_top['Qty Bins'] == 'Medium', 'Dish'].values):
  if(np.sum(matrix_df.loc[matrix_df['Qty Bins'] == 'High', 'Dish'].isin([dish])) == 1):
    to_promote_medium.append(dish)    
```

```
to_promote_low = []
for dish in list(goli_top.loc[goli_top['Qty Bins'] == 'Low', 'Dish'].values):
  if(np.sum(matrix_df.loc[matrix_df['Qty Bins'] == 'Medium', 'Dish'].isin([dish])) == 1):
    to_promote_low.append(dish)    
```

```
print(to_promote_medium)
print(to_promote_low)
```
